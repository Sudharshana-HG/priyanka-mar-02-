# Data type conversion

```python
data.info() 
```

```python
data['Month']=pd.to_numeric(data['Month'],errors='coerce') #[1,1.5,NAN] january NA
data['Temp C']=pd.to_numeric(data['Temp C'],errors='coerce')# coerce will introduce NA values for non numeric data in the columns
data['Weather']=data['Weather'].astype('category')           #data['Wind']=data['Wind'].astype('int64')
```

# Missing Values and Imputation

```python
data_cleaned3[data_cleaned3.isnull().any(axis=1)]
```

```python
data_cleaned3.isnull().sum()
```

```python
data_cleaned3.info()
```

```python
#Mean Imputationa
mean = data_cleaned3['Ozone'].mean()
print(mean)
```

```python
data_cleaned3['Ozone'] = data_cleaned3['Ozone'].fillna(mean)
```

```python
data_cleaned3
# Plot boxplot
```

```python
data_cleaned3['Weather']=data_cleaned3['Weather'].fillna(data_cleaned3['Weather'].mode().iloc[0])
```

```python
data_cleaned3.isnull().sum()
```

```python
solar=data_cleaned3['Solar'].mean()
print(solar)
```

```python
data_cleaned3['Solar']=data_cleaned3['Solar'].fillna(solar)
```

```python
data_cleaned3
```

```python
data_cleaned3.isnull().sum()
```

```python
month=data_cleaned3['Month'].mode()
print(month)
```

```python
data_cleaned3['Month']=data_cleaned3['Month'].fillna(data_cleaned3['Month'].mode().iloc[0])
data_cleaned3
```
